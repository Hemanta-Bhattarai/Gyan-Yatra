Now we have basic tools for data manipulation and data handeling. So let us learn to load a csv file (data file prepared from excel ) and do some manipulation and extraction.

On the folder we have iris.csv file. let us load the file

If you want to know about iris data set. Look at https://en.wikipedia.org/wiki/Iris_flower_data_set


```r
# load the csv file
# the iris.csv if loaded in data
data<-read.csv("iris.csv")
```

```r
# lets try to look what is the class of iris.csv file which is loaded in data
class(data)
```

```r
# its says the data is data.frame. As we know each data.frame has names which can be assesed by data.frame.name$var.name
# so lets look what are the heading of the data.frame data
names(data)  # to look the heading of the data frame
```

```r
# we can see there are 6 variables as given above
# To extract the data of Sepal.Lenth we must use data$Sepal.Length and lets put it in variable sepalLength
sepalLength<-data$Sepal.Length
sepalLength    # prints the data in under the variable name Sepal.Length
```

```r
# what are the types of data within the Petal.Width our guess is numerical
class(data$Petal.Width)
```

```r
# check what is types of variable in Species
class(data$Species)
```

```r
#lets print the data in Species
data$Species
```

```r
#finding mean of the sepal lenght of sepal.length of species setosa
# first extract data of sepal length for speices setosa
sepalLength.setosa<-data$Sepal.Length[data$Species=="setosa"]
mean(sepalLength.setosa)
```
