Here, I will load iris.csv file and find mean standard deviation, variance, correlation, co-variance etc of the data. I am only trying to show that different statistical tools that can be used.

```r
dataIris<-read.csv("iris.csv")
names(dataIris)
```

```r
#extracting the sepal and petal length and width for a species i.e versicolor
sepal_LengthVesi<-dataIris$Sepal.Length[dataIris$Species=="versicolor"]  #sepal length
petal_LengthVesi<-dataIris$Petal.Length[dataIris$Species=="versicolor"]  #petal lengh
sepal_WidthVesi<-dataIris$Sepal.Width[dataIris$Species=="versicolor"]    #sepal width
petal_WidthVesi<-dataIris$Petal.Width[dataIris$Species=="versicolor"]    #petal width

petalMeanLength<-mean(petal_LengthVesi)   #mean of the petal Length
message("Mean --->  ",petalMeanLength)


petalSdLength<-sd(petal_LengthVesi)      #standard deviation of the petal length
message("Standard Deviation--->  ",petalSdLength)


#lets check for corelation between the petal length and speal length
correlation<-cor(petal_LengthVesi,sepal_LengthVesi)
message("Coorelation between Petal length and sepal length ------ >  ",correlation)
```

```r
# to see normal statisics 
summary(petal_WidthVesi)
```

After showing some analysis with R let try to do some graphical plots

```r
#plot sepal width vrs speal length
plot(sepal_WidthVesi, sepal_LengthVesi,main="Width Vrs Length",xlab="Width",ylab="Length",col="red",lwd=10)
```

```r
# lets draw histogram of petal_width
hist(petal_WidthVesi,main="Histogram of petal width",col="green")
```

```r
#lets draw piechart of the species
dataSpecies<-dataIris$Species
dataSpeciesTable<-table(dataSpecies)   # table of the factors
pie(dataSpeciesTable,main="Pie Chart of Species",col=c("red","blue","green"))
```

```r
# try box plot speal_width of Vesi speices
boxplot(sepal_LengthVesi,main="Sepal Width",col="magenta")
```
