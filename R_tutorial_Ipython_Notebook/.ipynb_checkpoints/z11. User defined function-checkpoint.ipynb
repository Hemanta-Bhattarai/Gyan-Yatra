Example: Let us define our own function which return a value raised to other value i.e $a^b$

To do this let us say the function name we want to create is power. As we neet two numbers as input one is base(a) and other is power(b). So, there are two argument of the function. And the output of the funciton must be the value raised to a power ie $a^b$


    power<-function(a,b){       #power function defined with arguments a and b as variables like power(a,b)
    
        c=a^b                   # calculation of a raised to the power b
        
        return(c)               # Out put of the function is c
     }
     
     
if you want to find $2^5$ after defining the power function just call function power(2,5)

```r
 power<-function(a,b){       #power function defined with arguments a and b as variables like power(a,b)
    
        c=a^b                   # calculation of a raised to the power b
        
        return(c)               # Out put of the function is c
     }


result<-power(2,5)
print(result)
```
